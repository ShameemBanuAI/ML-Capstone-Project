# Working to find Outliers

# 1. Splitting of quan and qual columns

```python
quan=[]
qual=[]
for colName in df.columns:
    if (df[colName].dtypes)=='object':
        qual.append(colName)
    else:
        quan.append(colName)
```

```python
quan
```

```python
qual
```

# 2. Understanding the dataset deeper using Central Tendency, IQR ,Min & Max value

```python
descriptive=pd.DataFrame(index=["mean","median","mode","Q1:25","Q2:50","Q3:75","99th","Q4:100",
                               "IQR","1.5IQR","LesserRange","GreaterRange","MinValue","MaxValue"],columns=quan)
for colName in quan:
    descriptive[colName]["mean"]=df[colName].mean()
    descriptive[colName]["median"]=df[colName].median()
    descriptive[colName]["mode"]=df[colName].mode()[0]
    descriptive[colName]["Q1:25"]=df.describe()[colName]["25%"]
    descriptive[colName]["Q2:50"]=df.describe()[colName]["50%"]
    descriptive[colName]["Q3:75"]=df.describe()[colName]["75%"]
    descriptive[colName]["99th"]=np.percentile(df[colName],99)
    descriptive[colName]["Q4:100"]=df.describe()[colName]["max"]
    descriptive[colName]["IQR"]=descriptive[colName]["Q3:75"]- descriptive[colName]["Q1:25"]
    descriptive[colName]["1.5IQR"]=1.5 * descriptive[colName]["IQR"]
    descriptive[colName]["LesserRange"]= descriptive[colName]["Q1:25"] - descriptive[colName]["1.5IQR"]
    descriptive[colName]["GreaterRange"]= descriptive[colName]["Q3:75"] + descriptive[colName]["1.5IQR"]
    descriptive[colName]["MinValue"]=df[colName].min()
    descriptive[colName]["MaxValue"]=df[colName].max()
```

```python
descriptive
```

# 3. Checking for Outlier

```python
lesserOutlier=[]
greaterOutlier=[]
for colName in quan:
    if(descriptive[colName]["LesserRange"] > descriptive[colName]["MinValue"]):
        lesserOutlier.append(colName)
    if(descriptive[colName]["GreaterRange"] < descriptive[colName]["MaxValue"]):
        greaterOutlier.append(colName)
```

```python
lesserOutlier
```

```python
greaterOutlier
```

# 4. Replacing Ouliers

```python
for colName in lesserOutlier:
    df[colName][df[colName]<descriptive[colName]["LesserRange"]]=descriptive[colName]["LesserRange"]
for colName in greaterOutlier:
    df[colName][df[colName]>descriptive[colName]["GreaterRange"]]=descriptive[colName]["GreaterRange"]
```
