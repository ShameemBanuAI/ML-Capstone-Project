# 6)Test the Analysis of Variance between Time spent on fitness and Time spent on sleep at signifance level 5%. Make decision using Hypothesis Testing)

```python
import scipy.stats as stats

stats.f_oneway(df['Time spent on fitness'],df['Time spent on sleep'])
```

STATEMENT : Time spent on fitness and Time spent on sleep are significantly different 
NULL HYPO : There is not significant different in between Time spent on fitness , Time spent on sleep 
ALTER HYPO : There is significant different in between Time spent on fitness , Time spent on sleep 
SOLUTION : Rule : p<0.05 reject null hypo 0.00<0.05 true so accept alter hypo

# 7) What student missed more during pandemic 

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Assuming raw_df is your DataFrame containing the column you want to visualize
column_name = 'What you miss the most'  # Replace 'Cuisine' with your column name

# Extract the column from the DataFrame
column_data = df[column_name]

# Using word cloud the entire column is converted into single string. 
# Convert the column to a single string
text_data = ' '.join(map(str, column_data))

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text_data)

# Plot the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Turn off axis labels
plt.show()
```
